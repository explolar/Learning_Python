### Modules and Package
In Python,modules and package help in organise and reuse code.

```python
import math
math.sqrt(256)
```

```python
from math import sqrt,pi
print(sqrt(16))
print(pi)
```

```python
import numpy as np
r=np.array([1,2,3])
print(r)
```

```python
from math import *
print(sin(pi/2))
print(cos(0))
```

```python
from package.maths import add
print(add(5,10))

```
