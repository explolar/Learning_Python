```python
### Data Serialization
import json
data = {'name': 'Ankit', 'age': 22, 'city': 'kharagpur'}
json_string = json.dumps(data)  # Serialize to JSON string
print(json_string)
print(type(json_string))

data_back = json.loads(json_string)  # Deserialize back to Python object
print(data_back)
print(type(data_back))
```

```python
### csv Module
import csv
with open('data.csv', mode='w', newline='') as file: # Open file in write mode
    writer = csv.writer(file) # Create a csv writer object
    writer.writerow(['Name', 'Age', 'City']) # Write header
    writer.writerow(['Ankit', 22, 'Kharagpur']) # Write data row
    writer.writerow(['Rahul', 24, 'Delhi']) # Write data row

with open('data.csv', mode='r') as file: # Open file in read mode
    reader = csv.reader(file) # Create a csv reader object
    for row in reader:
        print(row) # Print each row
        
```

```python
## Date and Time Module
import datetime
from datetime import date, time, datetime, timedelta
now = datetime.now()
print("Current date and time:", now)
print("Current year:", now.year)
print("Current month:", now.month)
print("Current day:", now.day)
print("Current hour:", now.hour)
print("Current minute:", now.minute)
print("Current second:", now.second)

yesterday = now - timedelta(days=1) # Subtract one day
print("Yesterday's date and time:", yesterday)
tomorrow = now + timedelta(days=1) # Add one day
print("Tomorrow's date and time:", tomorrow)
```

```python
## time Module
import time
print("Current Time in seconds since Epoch:", time.time()) # Time since epoch
time.sleep(2) # Sleep for 2 seconds
print("Awake after 2 seconds")
print("Current Local Time:", time.localtime()) # Local time
print("Formatted Time:", time.strftime("%Y-%m-%d %H:%M:%S", time.localtime())) # Formatted time
```

```python
### Regular Expressions
import re

pattern = r'\bfoo\b'
text = 'foo bar baz fooqux foobar'
matches = re.findall(pattern, text)
print("Matches found:", matches)  

### Collections Module
from collections import Counter, defaultdict, namedtuple
# Counter
data = ['apple', 'banana', 'orange', 'apple', 'orange', 'banana', 'apple']
counter = Counter(data) 
print("Counter:", counter)
# defaultdict
dd = defaultdict(int)
dd['apple'] += 1    
dd['banana'] += 2
print("Defaultdict:", dd)
```
