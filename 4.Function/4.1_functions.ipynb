### Function
A function is a block of code that perform a specific task,helps in organizing,reusing code and improving readibility.

```python
## Syntax
def function_name(parameters):
    """docstring"""
    # function body
    return expression
```

### Why Functions?
Functions help in:
1. Code Reusability: Write once, use multiple times.
2. Modularity: Break down complex problems into smaller, manageable parts.
3. Readability: Clear structure and purpose of code.
4. Maintainability: Easier to update and fix bugs in isolated sections of code.

```python
num=10
if num%2==0:
    print(f"{num} is even")
else:
    print(f"{num} is odd")
```

```python
## But with Function
def even_odd(num):
    """checks whether a number is even or odd  """
    if num%2==0:
        return f"{num} is even"
    else:
        return f"{num} is odd"
```

```python
## Calling Function
even_odd(24)
```

```python
## Function with multiple parameters
def add(a,b):
    """adds two numbers"""
    c=a+b
    return c
```

```python
result=add(5,10)
print(result)
```

```python
## Default Parameters

def greet(name="Guest"):# Default parameter
    """Greets a person with their name"""
    return f"Hello, {name}!" 
print(greet("Ankit"))
print(greet() ) # Uses default parameter
```

```python
## Variable length Arguments
## Positional and Keyword Arguments

def print_no(num1,num2,num3,num4,num5):
    """Prints five numbers"""
    print(f"Numbers are: {num1}, {num2}, {num3}, {num4}, {num5}")
```

```python
def print_num(*args): # Variable length positional arguments
    """Prints all numbers passed as arguments"""
    for num in args:
        print(f"Number: {num}")
```

```python
print(print_no(1,2,3,4,5,6,7,8,9))
print(print_num(1,2,3,4,5,6,7,8,9))
```

```python
## Keyword Arguments
def print_info(**kwargs): # Variable length keyword arguments
    """Prints key-value pairs passed as arguments"""
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_info(name="Ankit", age=25, city="New York")
```

```python
def print_details(*args, **kwargs):
    """Prints positional and keyword arguments"""
    for item in args:
        print(f"position_Item: {item}")
    for key, value in kwargs.items():
        print(f"{key}: {value}")
```

```python
print_details(1,2,3, name="Ankit", age=25)
```
