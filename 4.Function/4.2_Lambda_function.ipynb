### Lambda Function ###
Lamda Function are small anonymous function defined using lamda keywords.they can have any no of argument but only one expression.They are commonly used for short operation or as a arguments to higher-ordered function.

```python
## Syntex
lambda arguments: expression
```

```python
def addition(a,b):
    return a+b

addition(2,3)
```

```python
add=lambda a,b:a+b
type(add)
print(add(5,7))
```

```python
def even(num):
    if num%2==0:
        return True
    
even(24)
```

```python
even1=lambda num:num%2==0
even(12)
```

```python
def add(x,y,z):
    return x+y+z
add(3,5,6)
```

```python
add1=lambda x,y,z:x+y+z
add(2,3,4)
```

```python
## map () - applies funtion to each collection in list
numbers=[1,2,3,4,5,6]
def square(number):
    return number**2

square(256)
```

```python
list(map(lambda x:x**2,numbers))
```
