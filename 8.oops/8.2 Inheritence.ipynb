## Inheritence ##
Inheritence is a fundamental concept in oops that allows a class to inherit attributes and methods from another class

```python
## Inheritence
## Inheritence is a fundamental concept in oops that allows a class to inherit attributes and methods from another class
class car:
    def __init__(self,windows,doors,engine):
        self.windows=windows
        self.doors=doors
        self.engine=engine
    def drive(self):
        print(f"The person is driving {self.engine} engine car with {self.windows} windows and {self.doors} doors")
```

```python
car1=car(4,4,"V6")
car1.drive()
```

```python
class Tesla(car):  
    def __init__(self, windows, doors, engine, autopilot):  ## adding new attribute autopilot
        super().__init__(windows, doors, engine)  ## inheriting from parent using super()
        self.has_autopilot = autopilot  ## renamed attribute to avoid conflict

    def enable_autopilot(self):  ## new method to enable autopilot
        if self.has_autopilot:  ## checking if autopilot is available
            print("Autopilot is enabled")
        else:
            print("This car does not have autopilot feature")
```

```python
tesla1 = Tesla(4, 5, "Electric", True)
tesla1.enable_autopilot()
tesla1.drive()
```

```python
## Multiple Inheritence ##
## Multiple Inheritence is a feature in oops where a class can inherit attributes and methods from more than one parent class
## Base Class 1
class Animal:
    def __init__(self, name):  # constructor to initialize name
        self.name = name

    def speak(self):  # method now actually speaks
        print(f"{self.name} makes a sound.")

    def sound(self):  # generic sound method
        print("Animal sound method")


class Pet:
    def __init__(self):  # optional constructor for Pet (since no attributes, can be empty)
        pass

    def play(self):
        print("Pet is playing.")


class Dog(Animal, Pet):
    def __init__(self, name, autopilot=False):  
        Animal.__init__(self, name)  # explicitly calling Animal constructor
        Pet.__init__(self)  # calling Pet constructor
        self.autopilot = autopilot  # example of Dog-specific attribute (optional)

    def sound(self):  # overriding base class method
        print(f"{self.name} says → Woof!")

    def bark(self):
        print("Woof! Woof!")
dog1 = Dog("Buddy")

dog1.speak()
dog1.play()
dog1.sound()
dog1.bark()
      
```
