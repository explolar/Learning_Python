```python
### Instance Variables and Methods
class Dog:
### Constructor to initialize instance variables
    def __init__(self,name,age):
        self.name=name ### Instance variable
        self.age=age ### Instance variable

##Creating an object of the class Dog
dog1=Dog("Buddy",3)
print(dog1)
print(dog1.name)
print(dog1.age)
    
```

```python
dog2=Dog("Max",5)
print(dog2.name)
print(dog2.age)
```

```python
dog3=Dog("Bella",2)
print(dog3.name)
print(dog3.age)
```

```python
## Define class with instance method
class Dog:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def bark(self): ### Instance method
        print(f"{self.name} says Woof!")
dog1=Dog("Buddy",3)
dog1.bark() ### Calling the instance method
dog2=Dog("Max",5)
dog2.bark() ### Calling the instance method

```

```python
### Modelling a bank account using OOP
## Define class
class BankAccount:
    def __init__(self,account_number,balance=0):
        self.account_number=account_number
        self.balance=balance
    def deposit(self,amount):
        self.balance +=amount
        print(f"Deposited {amount}. New balance is {self.balance}.")
    def withdraw(self,amount):
        if amount > self.balance:
            print("Insufficient funds.")
        else:
            self.balance -=amount
            print(f"Withdrew {amount}. New balance is {self.balance}.")
    def get_balance(self):
        return self.balance
    
# Create an object of the class BankAccount
account1=BankAccount("1234567890",1000) ## Initial balance is 1000
account1.deposit(500) ### Deposit 500
account1.withdraw(200) ### Withdraw 200
print(f"Current balance is {account1.get_balance()}.") ### Get current balance
```
