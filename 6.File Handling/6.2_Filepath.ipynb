```python
## Create a new directory
import os
new_directory = 'package_data'
os.makedirs(new_directory, exist_ok=True)
print(f"Directory '{new_directory}' created.")
```

```python
### List files in a directory
items = os.listdir('.')
print("Items in the current directory:",items)
```

```python
### Join paths to create a full file path
dir_name = 'package_data'
file_name = 'data.txt'
full_path = os.path.join(dir_name, file_name)
print("Full file path:", full_path)
```

```python
dir_name = 'package_data'
file_name = 'data.txt'
full_path = os.path.join(os.getcwd(),dir_name, file_name) # Get absolute path
print("Full file path:", full_path)
```

```python
path='example.txt'
if os.path.exists(path):
    print(f"The file '{path}' exists.")
else:
    print(f"The file '{path}' does not exist.")
```

```python
###  Checking if a path is a file or directory
path = 'example.txt'
if os.path.isfile(path):
    print(f"'{path}' is a file.")
elif os.path.isdir(path):
    print(f"'{path}' is a directory.")
else:
    print(f"'{path}' is neither a file nor a directory.")
```

```python
### Get absolute path of a file
relative_path = 'example.txt'
absolute_path = os.path.abspath(relative_path)
print("Absolute path:", absolute_path)
```
